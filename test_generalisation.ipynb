```python
%matplotlib inline 
import numpy as np
from keras.models import model_from_json 
from keras.datasets import mnist 
import matplotlib.pyplot as plot
from matplotlib import rcParams
import pickle
import sys
sys.path.insert(0, '/home/petra/pyRBF/')
from rbf import *


def probs_to_string(prob):
    # Pretty print vector of probabilities to string.
    res = ""
    maxi = np.argmax(prob)
    for p in range(len(prob)):
        #res += " %.2f" % prob[p]
        if p == maxi:
            res += "& {\\bf %.2f}" % prob[p] 
        else:
            res += "& %.2f " % prob[p] 
    return res + "\\\\"

FIG_NUM = 7

# Load saved individuals. 
X = np.load("adversary_inputs_against_rbf.npy")
X_mlp = np.load("adversary_inputs_matrix.npy")
X_ens = np.load("adversary_inputs_against_ensemble.npy")
X_cnn = np.load("adversary_inputs_against_cnn.npy")
X_svm_rbf = np.load("adversary_inputs_against_svm_rbf.npy")
X_svm_poly = np.load("adversary_inputs_against_svm_poly.npy")
X_svm_poly4 = np.load("adversary_inputs_against_svm_poly4.npy")
X_svm_sigmoid = np.load("adversary_inputs_against_svm_sigmoid.npy")
X_svm_linear = np.load("adversary_inputs_against_svm_linear.npy")
X_tree = np.load("adversary_inputs_against_tree.npy")

#  Load trained model. 
# mlp 
mlp = model_from_json(open("mlp.json").read()) 
mlp.load_weights("mlp_weights.h5")

# cnn
cnn = model_from_json(open("cnn.json").read())
cnn.load_weights("cnn_weights.h5")

# ensample
ensemble = []
for i in range(10):
    model = model_from_json(open("mlp.json").read())
    model.load_weights("mlp_weights_%s.h5" % i)
    ensemble.append(model)

# rbf 
rbf = pickle.load(open("rbf_mnist_4.pkl","rb"))

# svm 
svm_rbf = pickle.load(open("svm_rbf.pickle","rb"))
svm_poly = pickle.load(open("svm_poly.pickle","rb"))
svm_poly4 = pickle.load(open("svm_poly4.pickle","rb"))
svm_sigmoid = pickle.load(open("svm_sigmoid.pickle","rb"))
svm_linear = pickle.load(open("svm_linear.pickle","rb"))

tree = pickle.load(open("tree.pickle","rb"))

def eval_figure(X):
    """ Eval one figure with all networks. 
        parameter X : array 
                      input to eval 
                      
    """
    input = np.array([X])
    conv_input = np.array([X.reshape(1,28,28)])
    
    res = rbf.predict(input)
    print("RBF", probs_to_string(res[0])) 
    
    res = mlp.predict(input) 
    print("MLP", probs_to_string(res[0]))

    res = cnn.predict(conv_input)
    print("CNN", probs_to_string(res[0]))
    
    res = []
    for k in range(10):
        res.append( ensemble[k].predict(input)[0])
    res_mean = sum(res)
    res_mean /= 10 
    print("ENS", probs_to_string(res_mean)) 
    
    res = svm_rbf.predict_proba(input)
    print("SVM-rbf", probs_to_string(res[0])) 
    
    res = svm_poly.predict_proba(input)
    print("SVM-poly", probs_to_string(res[0]))
    
    res = svm_poly4.predict_proba(input)
    print("SVM-poly4", probs_to_string(res[0]))
    
    res = svm_sigmoid.predict_proba(input)
    print("SVM-sigmoid", probs_to_string(res[0]))
    
    res = svm_linear.predict_proba(input)
    print("SVM-linear", probs_to_string(res[0]))
    
    res = tree.predict_proba(input)
    print("Tree", probs_to_string(res[0]))
    

plot.figure(1)
#rcParams.update({'font.size': 8})

figures = [X[FIG_NUM], X_mlp[FIG_NUM], X_cnn[FIG_NUM], X_ens[FIG_NUM],
           X_svm_rbf[FIG_NUM], X_svm_poly[FIG_NUM], X_svm_poly4[FIG_NUM],
           X_svm_sigmoid[FIG_NUM], X_svm_linear[FIG_NUM], X_tree[FIG_NUM]]
names = ["RBF", "MLP", "CNN", "ENS",
        "SVM-RBF", "SVM-poly", "SVM-poly4", "SVM-sigmoid", "SVM-linear","Tree"]
for x, name in zip(figures, names):

    img = x.reshape(28,28)
    plot.imshow(img, interpolation="none", cmap=plot.cm.Greys)
    plot.title("Evolved against %s" % name)
    plot.savefig("%s_against_%s.eps" % (FIG_NUM, name), bbox_inches='tight')
    plot.show()
    
    eval_figure(x)
    print("------------------------------------------------------")
    
```
